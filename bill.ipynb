```python
from reportlab.platypus import SimpleDocTemplate, Table, Paragraph, TableStyle
from reportlab.lib import colors
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet
```

i copy past 1st line

```python
data=[["date","name","subcription","price(rs)"],
      ["3july2024","data analytic with python","6 weeks","Rs 10000/-"],
      ["3july2024","data analytic with python","6 month","35500/-"],
      ["sub total","","","45000/-"],
       ["discount","","","20000/-"]
       ,["total","","","25500/-"]
      ]
```

```python
pdf=SimpleDocTemplate("bill.pdf",pagesize=A4)
style=getSampleStyleSheet()
title=Paragraph("DUCAT",style["Heading1"])
```

```python
table = Table(data)
table.setStyle(TableStyle([
    ("BOX", (0, 0), (-1, -1), 1, colors.chocolate),
    ("GRID", (0, 0), (-1, -1), 1, colors.chocolate),
    ("BACKGROUND", (0, 0), (-1, 0), colors.chocolate),
    ("TEXTCOLOR", (0, 0), (-1, 0), colors.aqua),
    ("BACKGROUND", (0, 1), (-1, -1), colors.beige),
]))
pdf.build([title,table])
```
